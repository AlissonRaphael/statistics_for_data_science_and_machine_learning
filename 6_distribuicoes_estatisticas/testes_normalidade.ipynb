```python
from scipy import stats
import seaborn as sns
```

```python
dados_normais = stats.norm.rvs(size=1000)
```

```python
dados_nao_normais = stats.skewnorm.rvs(a=10, size=1000)
```

### Verificar se os dados s√£o normais

#### Histograma

```python
sns.histplot(dados_normais, kde=True)
```

```python
sns.histplot(dados_nao_normais, kde=True)
```

#### Quantile-quantile plot

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
qqplot(dados_normais, line='s')
```

```python
qqplot(dados_nao_normais, line='s')
```

#### Teste de Shapiro-Wilk

p-value √© usado para interpretar o teste estat√≠stico.
* p <= alpha: rejeita o teste de hip√≥tese, n√£o √© uma distribui√ß√£o normal.
* p > alpha: n√£o rejeita o teste de hip√≥tese, √© uma distribui√ß√£o normal.

```python
def teste(p, alpha=0.05):
    if p > alpha:
        print('√â uma distribui√ß√£o normal')
    else:
        print('N√£o √© uma distribui√ß√£o normal')
```

```python
from scipy.stats import shapiro
```

```python
_, p = shapiro(dados_normais)
_, p2 = shapiro(dados_nao_normais)
p
```

```python
teste(p, 0.05)
```

```python
teste(p2, 0.05)
```
